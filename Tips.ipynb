```python
import seaborn as sns

sns.violinplot(data=data, x="day", y="total_bill", hue="smoker"
   , split=True)
```

```python
g = sns.FacetGrid(data, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".1", fit_reg=False, x_jitter=.1)
```

```python
g = sns.FacetGrid(data, col="sex", height=4, aspect=.5)
g.map(sns.barplot, "smoker", "total_bill", order=["No", "Yes"])
```

```python
g = sns.FacetGrid(data, col="sex", height=4, aspect=.5)
g.map(sns.barplot, "size", "total_bill", order=[1,2,3,4,5,6])
```

```python
with sns.axes_style("white"):
    g = sns.FacetGrid(data, row="sex", col="smoker", margin_titles=True, height=2.5)
g.map(sns.scatterplot, "total_bill", "tip", color="#444888")
g.set_axis_labels("Total bill (US Dollars)", "Tip")
```
